# Extract the facial fetures from the images

```python
import facial_regions_detector as import_data
```

```python
def get_data():
    X, y_gender, y_smiling, image_files = import_data.extract_features_labels()
    print(X)
    Yg = np.array([y_gender, -(y_gender - 1)]).T
    Ys = np.array([y_smiling, -(y_smiling - 1)]).T
    tr_X = X[:3836] ; tr_Yg = Yg[:3836]; tr_Ys = Ys[:3836];
    te_X = X[3836:] ; te_Yg = Yg[3836:]; te_Ys = Ys[3836:]
    te_images = image_files[3836:]
    return tr_X, tr_Yg, tr_Ys, te_X, te_Yg, te_Ys, te_images
```

```python
training_images, training_labels_g, training_labels_s, test_images, test_labels_g, test_lables_s, test_images_string = get_data()
```

```python
test_labels_g
```

# Train a Simple Model Using the features Extracted

A simple neural network model will be defined and trained (for testing how accurate it can be). Nevertheless, this may not be the final model that I will be using.

```python
classifier = tf.keras.models.Sequential()
classifier.add(tf.keras.layers.Dense(units = 400, activation = 'relu', input_shape = (136,)))
classifier.add(tf.keras.layers.Dropout(0.2)) # Drop some neurons, along with their weights, to avoid overfitting the data

classifier.add(tf.keras.layers.Dense(units = 400, activation = 'relu'))
classifier.add(tf.keras.layers.Dropout(0.2))

classifier.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

The pixel values of the training images will now be scaled. 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
training_images_scaled = sc.fit_transform(training_images.reshape((3836, 136)))
```

```python
training_images_scaled
```

```python
classifier.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist = classifier.fit(training_images_scaled, new_training_labels_g, epochs = 100)
```

# Plotting the model loss and getting the testing accuracy

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

Let's now get the testing accuracy.

```python
# Reshape the images and get them scaled
test_images_scaled = sc.transform(test_images.reshape((959, 136)))
```

```python
predicted_test_labels_g = classifier.predict(test_images_scaled)
```

```python
predicted_test_labels_g
```

As can be seen, the predictions are numerical. We, however, want boolean predictions (True means male, False means female).

```python
predicted_test_labels_g = (predicted_test_labels_g > 0.5)
```

```python
predicted_test_labels_g
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(new_test_labels_g, predicted_test_labels_g)
```

```python
acc
```

The obtained accuracy was approximately 88.22%. This is a decent level of accuracy for the first model that is being tried.
