```python
# Extracted from https://www.kaggle.com/code/bmarcos/image-recognition-gender-detection-inceptionv3/notebook
def load_reshape_img(fname):
    img = load_img(fname)
    x = img_to_array(img)/255.
    x = x.reshape((1,) + x.shape)

    return x
```
